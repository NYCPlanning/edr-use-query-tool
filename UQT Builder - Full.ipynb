## Parse use group zoning district values


```python
use_groups_zoning_districts["Permitted Value"].value_counts()
```

```python
# def parse_naics_text(naics_text: str) -> str:
#     return naics_text.split("(")[1].removesuffix(")").replace("in", "").strip()
```

```python
PERMITTED_KEY = {
    "–": "Not permitted",
    "●": "Permitted",
    "♦": "Permitted with limitations",
    "*": "Permitted with limitations*",
    "○": "Special permit required",
    "S": "Size restriction",
    "P": "Additional conditions",
    "U": "Open use allowances",
}
PERMITTED_VALUE_COLUMNS = list(PERMITTED_KEY.values())
```

```python
def parse_permitted_value(row: pd.Series) -> list:
    for character in PERMITTED_KEY.keys():
        if character in row["Permitted Value"]:
            row[PERMITTED_KEY[character]] = True
        else:
            row[PERMITTED_KEY[character]] = False
    return row
```

```python
use_groups_zoning_districts = use_groups_zoning_districts.reindex(
    columns=[
        "Use Group",
        "Use Header",
        "Use Name",
        "Use NAICS Code",
        "NAICS Notes",
        "NAICS to subtract",
        "NAICS_Type",
        "Zoning District",
        "Permitted Value",
    ]
    + PERMITTED_VALUE_COLUMNS
)
use_groups_zoning_districts.head()
```

```python
use_groups_zoning_districts_flags = use_groups_zoning_districts.apply(parse_permitted_value, axis=1)

use_groups_zoning_districts_flags["Use NAICS Code"] = use_groups_zoning_districts_flags["Use NAICS Code"].astype(str)
use_groups_zoning_districts_flags['Use NAICS Code'] = use_groups_zoning_districts_flags['Use NAICS Code'].str.replace(' ', '')
use_groups_zoning_districts_flags["NAICS Notes"] = use_groups_zoning_districts_flags["NAICS Notes"].astype(str)
use_groups_zoning_districts_flags["NAICS to subtract"] = use_groups_zoning_districts_flags["NAICS to subtract"].astype(str).str.split('.').str[0]

use_groups_zoning_districts_flags.head()
```

```python
use_groups_zoning_districts_flags["Is Allowed"] = use_groups_zoning_districts_flags["Not permitted"].apply(lambda row: "No" if row else "Yes")

df_ug = use_groups_zoning_districts_flags.copy()

df_ug['Is Allowed'] = df_ug['Is Allowed'] + df_ug['Special permit required'].apply(lambda x: ', by special permit' if x else '')
df_ug['Is Allowed'] = df_ug['Is Allowed'] + df_ug['Size restriction'].apply(lambda x: ', subject to size restrictions' if x else '')
df_ug['Is Allowed'] = df_ug['Is Allowed'] + df_ug['Additional conditions'].apply(lambda x: ', subject to additional conditions' if x else '')
df_ug['Is Allowed'] = df_ug['Is Allowed'] + df_ug['Open use allowances'].apply(lambda x: ', subject to open use allowances' if x else '')
df_ug['Is Allowed'] = df_ug['Is Allowed'] + df_ug['Permitted with limitations'].apply(lambda x: ', with limited applicability' if x else '')
df_ug['Is Allowed'] = df_ug['Is Allowed'] + df_ug['Permitted with limitations*'].apply(lambda x: ', with limited applicability' if x else '')

df_ug.head()
```
