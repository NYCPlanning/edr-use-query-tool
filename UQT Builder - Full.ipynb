## Parse use group zoning district values


```python
use_groups_zoning_districts["Permitted Value"].value_counts()
```

```python
def parse_naics_text(naics_text: str) -> str:
    return naics_text.split("(")[1].removesuffix(")").replace("in", "").strip()
```

```python
# use_groups_zoning_districts["NAICS Code"] = use_groups_zoning_districts[
#     ["Uses (NAICS Code)"]
# ].map(parse_naics_text)
# use_groups_zoning_districts
```

```python
PERMITTED_KEY = {
    "–": "Not permitted",
    "●": "Permitted",
    "♦": "Permitted with limitations",
    "*": "Permitted with limitations*",
    "○": "Special permit required",
    "S": "Size restriction",
    "P": "Additional conditions",
    "U": "Open use allowances",
}
PERMITTED_VALUE_COLUMNS = list(PERMITTED_KEY.values())
```

```python
def parse_permitted_value(row: pd.Series) -> list:
    for character in PERMITTED_KEY.keys():
        if character in row["Permitted Value"]:
            row[PERMITTED_KEY[character]] = True
        else:
            row[PERMITTED_KEY[character]] = False
    return row
```

```python
use_groups_zoning_districts = use_groups_zoning_districts.reindex(
    columns=[
        "Use Group",
        "Use Header",
        "Use Name",
#         "Uses (NAICS Code)",
#         "NAICS Code",
        "Zoning District",
        "Permitted Value",
    ]
    + PERMITTED_VALUE_COLUMNS
)
use_groups_zoning_districts.head()
```

```python
use_groups_zoning_districts_flags = use_groups_zoning_districts.apply(parse_permitted_value, axis=1)
use_groups_zoning_districts_flags.head()
```

```python
use_groups_zoning_districts_flags["Is Allowed"] = use_groups_zoning_districts_flags.apply(lambda row: "No" if row["Not permitted"] else ("Yes*" if row[["Permitted with limitations", "Permitted with limitations*", "Special permit required", "Size restriction", "Additional conditions"]].any() else ("Yes" if row["Permitted"] else "No")), axis=1)
use_groups_zoning_districts_flags.head()
```
