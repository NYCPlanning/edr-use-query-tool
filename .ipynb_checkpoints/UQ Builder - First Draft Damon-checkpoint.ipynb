```python
import pandas as pd
from pathlib import Path
```

```python
RESOURCES_DIRECTORY = Path("./resources")
OUTPUT_DIRECTORY = Path("./output")
OUTPUT_DIRECTORY.mkdir(parents=True, exist_ok=True)
```

## Clean use groups exported from Excel


```python
use_groups = pd.read_csv(RESOURCES_DIRECTORY / "use_groups.csv")
use_groups
```

```python
use_groups_cleaned = (
    use_groups.replace({r"\s+$": "", r"^\s+": ""}, regex=True)
    .replace(r"\n", "", regex=True)
    .dropna()
)
use_groups_cleaned
```

```python
naics_description_columns = ["Uses Header", "Uses (NAICS Code)"]
other_columns = [
    col for col in use_groups_cleaned if col not in naics_description_columns
]

use_groups_cleaned[other_columns] = use_groups_cleaned[other_columns].applymap(
    lambda x: x.replace(" ", "").strip()
)
use_groups_cleaned
```

```python
use_groups_cleaned.to_csv(OUTPUT_DIRECTORY / "use_groups_cleaned.csv", index=False)
```

## Pivot use group zoning district columns


```python
id_columns = ["Use Group", "Uses Header", "Uses (NAICS Code)"]
use_groups_zoning_districts = use_groups_cleaned.melt(
    id_vars=id_columns, var_name="Zoning District", value_name="Permitted Value"
)
use_groups_zoning_districts
```

```python
use_groups_zoning_districts.to_csv(
    OUTPUT_DIRECTORY / "use_groups_zoning_districts.csv", index=False
)
```

## Parse use group zoning district values


```python
use_groups_zoning_districts["Permitted Value"].value_counts()
```

```python
def parse_naics_text(naics_text: str) -> str:
    return naics_text.split("(")[1].removesuffix(")").replace("in", "").strip()
```

```python
use_groups_zoning_districts["NAICS Code"] = use_groups_zoning_districts[
    ["Uses (NAICS Code)"]
].map(parse_naics_text)
use_groups_zoning_districts
```

```python
PERMITTED_KEY = {
    "–": "Not permitted",
    "●": "Permitted",
    "♦": "Permitted with limitations",
    "○": "Special permit required",
    "S": "Size restriction",
    "P": "Additional conditions",
    "U": "Open use allowances",
}
PERMITTED_VALUE_COLUMNS = list(PERMITTED_KEY.values())
```

```python
def parse_permitted_value(row: pd.Series) -> list:
    for character in PERMITTED_KEY.keys():
        if character in row["Permitted Value"]:
            row[PERMITTED_KEY[character]] = True
        else:
            row[PERMITTED_KEY[character]] = False
    return row
```

```python
use_groups_zoning_districts = use_groups_zoning_districts.reindex(
    columns=[
        "Use Group",
        "Uses Header",
        "Uses (NAICS Code)",
        "NAICS Code",
        "Zoning District",
        "Permitted Value",
    ]
    + PERMITTED_VALUE_COLUMNS
)
use_groups_zoning_districts
```

```python
use_groups_zoning_districts_flags = use_groups_zoning_districts.apply(parse_permitted_value, axis=1)
use_groups_zoning_districts_flags
```

```python
use_groups_zoning_districts_flags.to_csv(
    OUTPUT_DIRECTORY / "use_groups_zoning_districts_flags.csv", index=False
)
```

## Join NAICS code descriptions

```python
naics_codes = pd.read_csv(RESOURCES_DIRECTORY / "naics_codes.csv")
naics_codes
```

```python
naics_codes.columns
```

```python
naics_codes_cleaned = naics_codes.rename(columns={"2022 NAICS US   Code": "2022 NAICS US Code"}).drop(columns=["Seq. No."])
```
